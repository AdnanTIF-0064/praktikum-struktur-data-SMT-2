```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class DoubleLinkedList:
    def __init__(self):
        self.head = None

    def tambah_awal(self, data):
        new = Node(data)
        if self.head:
            self.head.prev = new
            new.next = self.head
        self.head = new

    def tampil(self):
        temp = self.head
        while temp:
            print(temp.data, end=" <-> ")
            temp = temp.next


    def sisip_tengah(self, kata_setelah, data_baru):
        temp = self.head
        while temp:
            if temp.data == kata_setelah:
                new = Node(data_baru)
                new.next = temp.next
                new.prev = temp
                if temp.next:
                    temp.next.prev = new
                temp.next = new
                return
            temp = temp.next

    def hapus(self, kata):
        temp = self.head
        while temp:
            if temp.data == kata:
                if temp.prev:
                    temp.prev.next = temp.next
                else:
                    self.head = temp.next
                if temp.next:
                    temp.next.prev = temp.prev
                return
            temp = temp.next


# ===== MAIN =====
dll = DoubleLinkedList()

dll.tambah_awal("C")
dll.tambah_awal("B")
dll.tambah_awal("A")

print("")
dll.tampil()

dll.sisip_tengah("B", "X")
print("")
dll.tampil()

dll.hapus("C")
print("")
dll.tampil()
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None

    def tambah(self, data):
        new = Node(data)
        if not self.head:
            self.head = new
            new.next = new
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = new
            new.next = self.head

    def tampil(self):
        if not self.head:
            return
        temp = self.head
        while True:
            print(temp.data, end=" -> ")
            temp = temp.next
            if temp == self.head:
                break


    def hapus_akhir(self):
        if not self.head:
            return
        if self.head.next == self.head:
            self.head = None
            return
        temp = self.head
        while temp.next.next != self.head:
            temp = temp.next
        temp.next = self.head


# ===== MAIN =====
cll = CircularLinkedList()

cll.tambah("1")
cll.tambah("2")
cll.tambah("3")

print("")
cll.tampil()

cll.tambah("4")
print("")
cll.tampil()

cll.hapus_akhir()
print("")
cll.tampil()
```
